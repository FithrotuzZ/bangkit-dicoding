### Pertanyaan 2: Bagaimana performa penjualan seiring berjalannya waktu dilihat dari setiap bulannya?

```python
monthly = all_data.resample(rule='M', on='order_approved_at').agg({
    "order_id": "nunique",
})
monthly.index = monthly.index.strftime('%B') #mengubah format order_approved_at menjadi Tahun-Bulan
monthly = monthly.reset_index()
monthly.rename(columns={
    "order_id": "order_count",
}, inplace=True)
monthly.head()
```

```python
#mengurutkan jumlah pesanan, kemudian menghapus duplikat berdasarkan tanggal persetujuan pesanan
monthly = monthly.sort_values('order_count').drop_duplicates('order_approved_at', keep='last')
```

```python
month_mapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

monthly["month_numeric"] = monthly["order_approved_at"].map(month_mapping)
monthly = monthly.sort_values("month_numeric")
monthly = monthly.drop("month_numeric", axis=1)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(
    monthly["order_approved_at"],
    monthly["order_count"],
    marker='o',
    linewidth=2,
    color="#4b7bb5"
)
plt.title("Jumlah Penjualan Setiap Bulan (2018)", loc="center", fontsize=15)
plt.xticks(fontsize=12, rotation=25)
plt.yticks(fontsize=12)
plt.show()
```

**Insight:** Pada grafik diatas dapat dilihat bahwa terjadi penurunan signifikan dari penjualan bulan Agustus ke bulan September dan juga penjualan mengalami kenaikan yang signifikan dari bulan Oktober ke bulan November.

###Pertanyaan 3: Lokasi geografis manakah yang memiliki pelanggan terbanyak?

```python
bystate = all_data.groupby(by="customer_state").customer_id.nunique().reset_index()
bystate.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)
bystate.head()
```

```python
plt.figure(figsize=(10, 5))

most_common_state = bystate.loc[bystate['customer_count'].idxmax(), 'customer_state']

bystate = bystate.sort_values(by='customer_count', ascending=False)

sns.barplot(x='customer_state',
            y='customer_count',
            data=bystate,
            palette=["#4b7bb5" if state == most_common_state else "#D3D3D3" for state in bystate['customer_state']]
            )

plt.title("Jumlah Customers Setiap State", fontsize=15)
plt.xlabel("State")
plt.ylabel("Jumlah Customers")
plt.xticks(fontsize=10)
plt.show()
```

**Insight:** Pada grafik diatas customer terbanyak berdasarkan State yaitu SP

```python
bycity = all_data['customer_city'].value_counts().head(10)

plt.figure(figsize=(8, 4))

most_common_city = bycity.idxmax()

bycity = bycity.sort_values(ascending=False)

sns.barplot(x=bycity.index,
            y=bycity.values,
            palette=["#4b7bb5" if city == most_common_city else "#D3D3D3" for city in bycity.index]
            )

plt.title("Jumlah Customers Setiap City", fontsize=15)
plt.xlabel("City")
plt.ylabel("Jumlah Customers")
plt.xticks(rotation=15, fontsize=10)
plt.show()
```

**Insight:** Pada grafik diatas customer terbanyak berdasarkan City adalah Sao Paulo

## Analisis RFM

```python
all_data['order_purchase_timestamp'].max()
```

```python
all_data['order_purchase_timestamp'].min()
```

```python
#analogi now = 20/10/2018
import datetime as dt
now=dt.datetime(2018,10,20)
```

```python
all_data['order_purchase_timestamp'] = pd.to_datetime(all_data['order_purchase_timestamp'])

# Group by 'customer_id' and calculate Recency, Frequency, and Monetary
recency = (now - all_data.groupby('customer_id')['order_purchase_timestamp'].max()).dt.days
frequency = all_data.groupby('customer_id')['order_id'].count()
monetary = all_data.groupby('customer_id')['price'].sum()

# Create a new DataFrame with the calculated metrics
rfm = pd.DataFrame({
    'customer_id': recency.index,
    'Recency': recency.values,
    'Frequency': frequency.values,
    'Monetary': monetary.values
})
```

```python
col_list = ['customer_id','Recency','Frequency','Monetary']
rfm.columns = col_list
```

```python
rfm
```

```python
rfm.sort_values(by='Monetary',ascending=False) #sort rfm berdasarkan monetary
```

```python
rfm.sort_values(by='Recency',ascending=False) #sort rfm berdasarkan recency
```

```python
rfm.sort_values(by='Frequency',ascending=False) #sort rfm berdasarkan frequency
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#4b7bb5", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(y="Recency", x="customer_id", data=rfm.sort_values(by="Recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel("customer_id")
ax[0].set_title("Berdasarkan Recency", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=75)
ax[0].set_xticks([]) #beri comentar pada baris ini jika ingin melihat customer nya by id


sns.barplot(y="Frequency", x="customer_id", data=rfm.sort_values(by="Frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel('customer_id')
ax[1].set_title("Berdasarkan Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
ax[1].set_xticklabels(ax[0].get_xticklabels(), rotation=75)
ax[1].set_xticks([]) #beri comentar pada baris ini jika ingin melihat customer nya by id


sns.barplot(y="Monetary", x="customer_id", data=rfm.sort_values(by="Monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel('customer_id')
ax[2].set_title("Berdasarkan Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
ax[2].set_xticklabels(ax[0].get_xticklabels(), rotation=70)
ax[2].set_xticks([]) #beri comentar pada baris ini jika ingin melihat customer nya by id


plt.suptitle("Customer Terbaik dengan Analisis RFM Berdasarkan (customer_id)", fontsize=20)
plt.show()
```
